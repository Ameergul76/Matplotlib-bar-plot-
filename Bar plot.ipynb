```python
x=["python","c++","c","java"]
y=[80,70,75,69]

plt.xlabel("language",fontsize=15)
plt.ylabel("Ranking",fontsize=15)
plt.title("Pyfusion")
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,align="edge",edgecolor="r",linewidth=3,linestyle=":",alpha=0.5)
plt.show()
```
